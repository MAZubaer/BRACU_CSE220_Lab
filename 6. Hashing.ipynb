```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    if self.search_hashtable(s) == 'Found':
      print(s,'Already Inserted. Cannot reinsert.')
      print('===============================')
      return
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(key: {head.key}, value: {head.value})', end = '-->')
        head = head.next
      print('None')
      print()
      idx += 1

  def Double_ascii_sum(self, st="", sum=0):
    if len(st)==0:
      return sum
    elif len(st)<2:
      temp=str(ord(st[0]))
      sum+=int(temp)
      st=""
      return self.Double_ascii_sum(st, sum)
    elif len(st)>=2:
      temp=str(ord(st[0]))+str(ord(st[1]))
      sum+=int(temp)
      return self.Double_ascii_sum(st[2:], sum)

  def __hash_function(self, key):
    if len(key)%2!=0:
      key+="N"
    index= self.Double_ascii_sum(key)%len(key)
    return index

  def search_hashtable(self, s, count=0):
    if self.ht[count]==None:
      pass
    elif self.ht[count].next==None:
      if self.ht[count].key==s[0] and self.ht[count].value==s[1]:
        return "Found"
      else:
        pass
    elif self.ht[count].next!=None:
      temp=self.ht[count]
      flag= "False"
      while temp!=None:
        if temp.key==s[0] and temp.value==s[1]:
          flag= "True"
          break
        temp=temp.next
      if flag=="True":
        return "Found"
      else:
        pass
    if count<(len(self.ht)-1):
      count+=1
      return self.search_hashtable(s, count)
    else:
      return "Not Found"


```

Implement the __hash_function() and insert() methods

```python
class Node:
  def __init__(self, value=None, next = None):
    self.value = value
    self.next = next

class HashTable:
  def __init__(self, length):
    n = Node()
    self.ht = [n] * length
    self.length = length

  def show(self):
    count = 0
    for i in self.ht:
      temp = i
      print(count, end=" ")
      while temp!=None:
        print (temp.value, end="-->")
        temp = temp.next
      count += 1
      print()

  def even_ascii_sum(self, key, count=0):
    if count+2>(len(key)+1):
      return 0
    else:
      return ord(key[count])+self.even_ascii_sum(key,count+2)

  def odd_ascii_sum(self, key, count=0):
    if count==0:
      return self.odd_ascii_sum(key, count+1)
    elif count+2>(len(key)+1):
      return 0
    else:
      return ord(key[count])+self.odd_ascii_sum(key, count+2)

  def __hash_function(self, key):
    if len(key)%2==0:
      index=self.even_ascii_sum(key)%len(self.ht)
    else:
      index=self.odd_ascii_sum(key)%len(self.ht)
    return index

  def insert(self, key, value):
    idx=self.__hash_function(key)
    if self.ht[idx].value==None:
      n=Node((key, value), None)
      self.ht[idx]=n
    else:
      if self.ht[idx].next==None:
        if self.ht[idx].value[1]>=value:
          n=Node((key, value), None)
          self.ht[idx].next=n
        else:
          n=Node((key, value), None)
          temp=self.ht[idx]
          n.next=temp
          self.ht[idx]=n
      else:
        temp=self.ht[idx]
        while temp!=None:
          if temp.next!=None:
            if temp.value[1]>=value>=temp.next.value[1]:
              temp1=temp
              temp2=temp.next
              n=Node((key, value), None)
              temp1.next=n
              n.next=temp2
              break
            elif temp.value[1]<value:
              n=Node((key, value), None)
              temp=self.ht[idx]
              n.next=temp
              self.ht[idx]=n
              break
          else:
            n=Node((key, value), None)
            temp.next=n
          temp=temp.next
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1



  def __hash_function(self, key):
    idx=(key+3)%6
    return idx

  def remove_list(self, head, idx):
    count=0
    temp=head
    while temp!=None:
      count+=1
      temp=temp.next
    if idx==0:
      head=head.next
    elif idx>=1 and idx<count:
      temp=head
      obj=None
      ct=0
      while temp!=None:
        if ct==idx-1:
          obj=temp
          break
        temp=temp.next
        ct+=1
      removed_node=obj.next
      obj.next=removed_node.next
    return head


  def remove(self, key, count=0):
    if self.ht[count]==None:
      pass
    else:
      temp=self.ht[count]
      idx=0
      while temp!=None:
        if temp.key==key:
          self.ht[count]=self.remove_list(self.ht[count],idx)
          idx-=1
        idx+=1
        temp=temp.next

    if count<(len(self.ht)-1):
      count+=1
      return self.remove(key, count)
    else:
      return
```
